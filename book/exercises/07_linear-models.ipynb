```r
#INSERT CODE HERE
lm_fit <- lm(Flanker_Unadj ~ FS_Total_GM_Vol, data = d1)
summary(lm_fit)
coef(lm_fit)
confint(lm_fit)
```

> *Grey matter volume is significantly associated with Flanker Task performance, as the p-value is <0.05 (3.18e-12). However, the R² = 0.0428, which is quite small. This indicates that grey matter volume explains only a very small part of the variance in Flanker performance.*

## 4. Plotting (2 points)

Use `ggplot` to plot the `FS_Total_GM_Vol` variable (x axis) against the `Flanker_Unadj` variable (y axis), as well as the regression line with confidence intervals on the regrssion line. Qualitatively describe what you see.

```r
#INSERT CODE HERE
ggplot(d1, aes(x = FS_Total_GM_Vol, y = Flanker_Unadj)) +
  geom_point() +
  geom_smooth(method = "lm", se = TRUE)
```

Although the regression line goes upward, the data points are widely dispersed around the line and the slope isn't very positive, indicating a weak association. The confidence interval is relatively narrow due to the large sample size, suggesting the slope estimate is precise despite the small effect size.*
